## Testing the knots the HKL paper

This is the paper Hedden, Kirk, Livingston, "Non-slice linear combinations of algebraic knots".
The knots are of the form
$$ J(p,q,q_{1},q_{2}) = T(p,q;p,q_{1}) \# -T(p,q_{1}) \# -T(p,q;p,q_{2}) \# T(p,q_{2}).$$

```python
def create_J_knot(p,q,q1,q2):
    knot_desc = [
        (1,
         [
             (p, q),
             (p, q1)
         ]
        ),
        (-1,
         [
             (p, q1)
         ]
        ),
        (-1,
         [
             (p, q),
             (p, q2)
         ]
        ),
        (1,
         [(p, q2)]
        )
    ]
    return GeneralizedAlgebraicKnot(knot_desc)
```

```python
from sage.all import gcd


def generate_test_cases(p, q, limit=50):
    test_cases = [(p, q, q1, q2) for q1 in range(max(p, q) + 1, limit) for q2 in range(q1 + 1, limit) if gcd(q1, p) == 1 and gcd(q2, p) == 1]
    return test_cases
```

```python
%%ipytest -v -W ignore::DeprecationWarning

test_cases = generate_test_cases(2,3)

@pytest.mark.parametrize('knot_description', test_cases)

def test_case_2_3(knot_description):
    K = create_J_knot(*knot_description)
    sigma = K.signature()
    assert sigma.is_zero_everywhere()
```
