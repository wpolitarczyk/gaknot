```python
%%ipytest -vv -W ignore::DeprecationWarning

@pytest.mark.parametrize("case_idx", range(10))
def test_slicen_random(case_idx):
    # Varied number of layers from 1 to 4
    num_layers = (case_idx % 4) + 1
    
    # 1. Generate a random iterated torus knot K with constrained p, q for performance
    K_desc = gen_iterated(size=num_layers, size_range=40)
    
    # 2. Compute J = slicen(K)
    J = slicen(K_desc)
    
    # 3. Form the connected sum L = K # J
    L_desc = [(1, K_desc)] + J
    
    # 4. Create the GeneralizedAlgebraicKnot object
    K_total = GeneralizedAlgebraicKnot(L_desc)
    
    # 5. Verify the signature is zero everywhere
    sig = K_total.signature()
    assert sig.is_zero_everywhere(), f"Case {case_idx} failed: Signature is not zero everywhere.\nK: {K_desc}\nJ: {J}"
```
